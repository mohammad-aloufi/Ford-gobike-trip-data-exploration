## Investigation Overview

We'll do a bit of investigation on this dataset. It's about bike trips where you rent a bike for a time and use it to go from place to another without worrying about buying an actual bike.
We will see how the users use the service, who uses it the most? and when?


## Dataset Overview

The dataset we have contains the data about all trips that are done in February

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe

df =pd.read_csv('fordgobike-tripdata.csv')
```

## Gender usage

Males use the service more than females? That's not surprising to me.

```python
#Gender graph
sns.countplot(data=df, x ='member_gender', color ='black');
plt.title('Gender_usage');
plt.xlabel('Gender');
plt.ylabel('Count');
```

## User type usage

subscribers use the survice more than customers. That's surprising to say the least. Even though, Customers and Subscribers seems like they go on trips on Thursdays the most.

```python
#Week days and user type
sns.countplot(data =df, x ='start_day', hue ='user_type');
plt.title('Usertype_usage');
plt.xlabel('Day');
plt.ylabel('Count');
```

## Week day, user type, and trip duration relationship

Customers take longer trips than subscribers in average.

```python
sns.pointplot(data =df, x ='start_day', y ='duration_min', hue ='user_type');
plt.title('Weekday_usertype_duration');
plt.xlabel('Day');
plt.ylabel('Duration');
```
